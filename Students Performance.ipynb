```python
duplicates=df.duplicated().sum()
print(f"\nDuplicate Rows: {duplicates}")
if duplicates >0:
    df.drop_duplicates(inplace=True)
```

## Encoding Categorical Variables

```python
##Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['gender_encoded'] = le.fit_transform(df['gender'])
```

```python
## One Hot Encoding

df = pd.get_dummies(df, columns=['lunch', 'race/ethnicity','test preparation course'], drop_first=True)
```

```python
## Ordinal encoding

from sklearn.preprocessing import OrdinalEncoder

edu_order = [['some high school','high school','some college','associate\'s degree', 
              'bachelor\'s degree', 'master\'s degree']]
oe = OrdinalEncoder(categories=edu_order)
df['parental_edu_encoded'] = oe.fit_transform(df[['parental level of education']])
```
